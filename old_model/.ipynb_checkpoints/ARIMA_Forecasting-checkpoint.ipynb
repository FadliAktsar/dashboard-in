# Stationary Test

```python
result = adfuller(train_df['volume'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:')
for key, value in result[4].items():
    print(f'\t{key}: {value}')
```

### ADF Statistic: -6.457878419589361
The ADF Statistic is a negative value. It is a test statistic that is more negative for a more stationary time series. In this case, the value of the ADF Statistic (-6.457878419589361) indicates that the time series is likely to be stationary.

### p-value: 1.4679525912915281e-08 (approximately 0.000000001467)
The p-value is the probability of observing a test statistic as extreme as the one calculated from the data, assuming that the null hypothesis is true. In this case, the extremely low p-value (close to zero) suggests strong evidence against the null hypothesis of non-stationarity. When the p-value is less than the chosen significance level (commonly 0.05), we reject the null hypothesis and conclude that the data is likely stationary.

### Critical Values:
The critical values are thresholds that serve as reference points for determining the significance of the ADF Statistic. They correspond to different levels of significance (1%, 5%, and 10%) at which the null hypothesis can be rejected. In this case, all the critical values are more negative than the ADF Statistic, further supporting the rejection of the null hypothesis and indicating that the data is likely stationary.

Based on the ADF test results, we have strong evidence to reject the null hypothesis of non-stationarity. Therefore, we can conclude that the time series data is stationary. This is good news because many time series forecasting methods, including ARIMA, require the data to be stationary to provide accurate and meaningful predictions.

```python
def test_stationary(timeseries):
    timeseries = timeseries.to_frame()

    #Determine rolling statistics
    movingAverage = timeseries.rolling(window=12).mean()
    movingSTD = timeseries.rolling(window=12).std()

    #Plot rolling statistics
    orig = plt.plot(timeseries, color='blue', label='Original')
    mean = plt.plot(movingAverage,color='red', label='Rolling Mean')
    std = plt.plot(movingSTD, color='black', label='Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standart Deviation')
    plt.show(block=False)

    #Perform Dickey-Fuller
    print('Results of Dickey Fuller Test')
    dftest = adfuller(timeseries['volume'], autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observation Used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationary(train_df['volume'])
```

# Ploting ACF dan PACF untuk melihat nilai p, d, q

```python
#ACF & PACF plots

lag_acf = acf(train_df['volume'], nlags=12)
lag_pacf = pacf(train_df['volume'], nlags=12, method='ols')

# Plot ACF and PACF
plt.figure(figsize=(12, 6))

#Plot ACF
plt.subplot(121)
plt.stem(range(len(lag_acf)), lag_acf, use_line_collection=True)
#plt.plot(lag_acf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(train_df)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(train_df)), linestyle='--', color='gray')
plt.title('Autocorrelation Function')

#Plot PACF
plt.subplot(122)
plt.stem(range(len(lag_pacf)), lag_pacf, use_line_collection=True)
#plt.plot(lag_pacf)
plt.axhline(y=0, linestyle='--', color='gray')
plt.axhline(y=-1.96/np.sqrt(len(train_df)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(train_df)), linestyle='--', color='gray')
plt.title('Partial Autocorrelation Function')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Data (assuming it is loaded into a DataFrame named 'df')
df['year_month'] = pd.to_datetime(df['year_month'], format='%d/%m/%Y')
df.set_index('year_month', inplace=True)
# Split the data into train and test (assuming you have the last 12 months for testing)
train_df = df.iloc[:60]
test_df = df.iloc[48:]  # Predict months 48 to 60 (the last 12 months)
```

## Model Fitting

```python
# Fit the ARIMA model
p, d, q = 5, 0, 3  # Choose appropriate ARIMA parameters (can be obtained from auto_arima)
model = sm.tsa.ARIMA(train_df['volume'], order=(p, d, q))
arima_result = model.fit()
```

```python
arima_result.summary()
```

# Peramalan Data

```python
# Forecast future values (predict months 48 to 60)

forecast_results = arima_result.get_prediction(start='2015-01-01', end='2015-12-01', dynamic=False)
forecast_values = forecast_results.predicted_mean
conf_int = forecast_results.conf_int()
forecast_df = pd.DataFrame({'Forecast': forecast_values}, index=test_df.index)

# Merge forecasted and test values based on the index
merged_df = pd.concat([forecast_df, test_df], axis=1)
```

# Evaluasi Data Peramalan

```python
# Calculate RMSE and MAE for the predicted period
rmse = mean_squared_error(merged_df['volume'], merged_df['Forecast'], squared=False)
mae = mean_absolute_error(merged_df['volume'], merged_df['Forecast'])

print('Root Mean Squared Error (RMSE):', rmse)
print('Mean Absolute Error (MAE):', mae)

```
